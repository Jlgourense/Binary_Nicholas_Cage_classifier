```python
def pics_url_page(query,page):
    web="https://www.google.com/search"
    payload={"q":'{}'.format(query),"tbm":"isch","start":"{}".format(str(int(page)*20))}
    logger.debug('payload:{}'.format(payload))
    r=requests.get(web,params=payload)
    soup=BeautifulSoup(r.content,'html.parser')
    target=soup.find_all('img')
    pics=[x['src'] for x in target]
    return pics
#pics=pics_url_page('face',1)
```

# Download picture


```python
import shutil

def download_picture(url,name):
    r=requests.get(url,stream=True)
    with open("./images_notnic/nic_{}".format(name),'wb') as f:
        r.raw.decode_content = True
        shutil.copyfileobj(r.raw, f) 
        
```

## Downloading

```python
import time
import numpy as np

n_pages=50
for n in range(n_pages):
    pics=pics_url_page('face',n)
    for i,pic in enumerate(pics):
        download_picture(pic,"{}".format(str(i+n*20)))
    time.sleep(np.random.rand()*2)
```
